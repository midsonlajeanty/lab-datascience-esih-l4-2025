```python
import re
import csv
import time
import requests
from bs4 import BeautifulSoup

host = 'https://haitiwonderland.com'
path = "/news"
url = host + path

list_article = []

for i in range(50):

    print ("Requete #", i)

    time.sleep(1)

    reponse = requests.get(url)

    if(reponse.status_code != 200):
        exit()
        
    soup = BeautifulSoup(reponse.text, "html.parser")

    article_boxes = soup.find_all('div', class_='blog-item-container-box')

    for article_box in article_boxes:
        titre = str.strip(article_box.find('h2').text).replace("\n", "")
        categorie = str.strip(article_box.find('span', class_='blog-item-category').text).replace("\n", "")
        date = str.strip(article_box.find('span', class_='date-style').text).replace("\n", "")
        auteur = str.strip(re.sub(r'\s*\|.*', '', str(article_box.find('h6').text).replace("Par ", ""))).replace("\n", "")
        lien = str.strip(article_box.find('a').attrs['href']).replace("\n", "")
        image_url = host + str.strip(re.sub(r'background-image:url\(["\']?(.*?)["\']?\);', r'\1', article_box.find('div', class_="blog-item-img").attrs['style'])).replace("\n", "")

        article_dict = {
            "titre" : titre,
            "categorie": categorie,
            "date": date,
            'auteur': auteur,
            "lien": lien,
            'image_url': image_url
        }

        exists = list(filter(lambda x: x['titre'] == titre, list_article))

        if len(exists) <= 0:
            list_article.append(article_dict)



with open("haiti-wonder-land.csv", 'w', encoding="utf-8") as f:
    writer = csv.DictWriter(
        f=f, 
        lineterminator="\n",
        fieldnames=['titre', 'categorie', 'date', 'auteur', 'lien', 'image_url']
    )
    writer.writeheader()
    writer.writerows(list_article)
```
