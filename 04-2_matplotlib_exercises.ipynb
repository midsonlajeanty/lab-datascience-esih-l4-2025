```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5, 9]
y = [10, 20, 25, 30, 40, 35]

plt.plot(x, y, marker='o', linestyle='--', color='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')

plt.show()
```

```python
plt.figure(figsize=(6, 5))
plt.plot(x, y, marker='o', linestyle='--', color='g')
plt.xlabel('X-axis', fontsize=18)
plt.ylabel('Y-axis', fontsize=14)

plt.legend(["Ligne 1"], loc="upper right")
plt.grid(True)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 8, 27, 64, 125]
y3 = [2, 4, 6, 8, 10]


plt.plot(x, y1, 'r-o', label="Carr")
plt.plot(x, y2, 'b-+', label="Cube")
plt.plot(x, y3, 'g--*',label="Double")

plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)

plt.title("Description", fontsize="18")
plt.legend()
plt.grid(True)


plt.show()
```

```python
data = [10, 20, 20, 30, 30, 30, 40, 50, 50, 60]

plt.hist(data, color='r', edgecolor="g")
plt.xlabel("Valeurs")
plt.ylabel("Frquence")
plt.title("Histogramme")

plt.show()
```

```python
categories = ["A", "B", "C", "D"]
valeurs = [5, 7, 3, 8]

plt.bar(categories, valeurs, color=['red', 'blue', 'green', 'orange'])
plt.xlabel("Catgories")
plt.ylabel("Valeurs")
plt.title("Diagramme en barres")
plt.show()
```

```python
labels = ["A", "B", "C", "D", "E"]
sizes = [30, 20, 40, 10, 90]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'lightblue', 'red', 'green', "orange"])
plt.title("Diagramme circulaire")

# plt.savefig('pie-chart.png')
plt.show()
```
